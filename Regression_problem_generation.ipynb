```python
#DGP1-1
#one break
for sigma_u in [0.5, 1.0, 1.5]:
    for T in [100,200,500]:
        df = pd.DataFrame()
        fname = 'DGP-1_%s_%s'%(sigma_u,T)
        for i in tqdm(range(500)):
            np.random.seed(i+T*10)
            X1 = np.random.normal(0,1,T)
            res = np.random.normal(0,sigma_u,T)
            y0 = 0*X1+res 
            y1 = 1*X1+res
            y = np.hstack([y0[0:int(T/2)],y1[int(T/2):T]])
            df['x_%s'%i] = X1
            df['y_%s'%i] = y
        df.to_csv('/Users/username/Dropbox/MIP_regr/MIP_regr_500/one_break/%s.csv'%fname)
```

```python
#DGP2
#one break
for sigma_u in [0.5, 1.0, 1.5]:
    for T in [100,200,500]:
        df = pd.DataFrame()
        fname = 'DGP-2_%s_%s'%(sigma_u,T)
        for i in tqdm(range(500)):
            np.random.seed(i+T*10+10000)
            
            X1 = np.random.normal(0,1,T)
            eps = np.random.normal(0,np.sqrt(0.75),T+10)
            v = []
            v += [eps[0]]
            for j in range(1,T+10):
                v += [0.5*v[j-1] + eps[j]]
            u = np.array(v[10:])
            u*=sigma_u

            y0 = 0*X1+u 
            y1 = 1*X1+u

            y = np.hstack([y0[0:T//2],y1[T//2:T]])
            df['x_%s'%i] = X1
            df['y_%s'%i] = y
        df.to_csv('/Users/username/Dropbox/MIP_regr/MIP_regr_500/one_break/%s.csv'%fname)
```

```python
#DGP4
#one break
for sigma_u in [0.5, 1.0, 1.5]:
    for T in [100,200,500]:
        df = pd.DataFrame()
        fname = 'DGP-4_%s_%s'%(sigma_u,T)
        for i in tqdm(range(500)):
            np.random.seed(i+10*T+30000)
            
            eps = np.random.normal(0,np.sqrt(0.75),T+10)

            X1 = []
            X1 += [eps[0]]
            for j in range(1,T+10):
                X1 += [0.5*X1[j-1] + eps[j]]
        
            
            eps_1 = np.random.normal(0,1,T+10)
            u = []
            h = []
            sigma = sigma_u

            u+=[sigma*eps_1[0]]
            h+=[0.05 + 0.05*u[0]**2]

            for j in range(1,T+10):
                h += [0.05 + 0.05*u[j-1]**2 + 0.9*h[j-1]]
                u += [sigma*np.sqrt(h[j])*eps_1[j] ]

            X1 = np.array(X1[10:])
            u = np.array(u[10:])
            y0 = 0*X1+u 
            y1 = 1*X1+u


            y = np.hstack([y0[0:T//2],y1[T//2:T]])
            df['x_%s'%i] = X1
            df['y_%s'%i] = y
        df.to_csv('/Users/username/Dropbox/MIP_regr/MIP_regr_500/one_break/%s.csv'%fname)
```

```python
#DGP5
#one break
for sigma_u in [0.5, 1.0, 1.5]:
    for T in [100,200,500]:
        df = pd.DataFrame()
        fname = 'DGP-5_%s_%s'%(sigma_u,T)
        for i in tqdm(range(500)):
            np.random.seed(i+T*10+40000)
            eps = np.random.normal(0,np.sqrt(0.75),T+10)
            X1 = []
            X1 += [eps[0]]
            for j in range(1,T+10):
                X1 += [0.5*X1[j-1] + eps[j]]

            eps_1 = np.random.normal(0,np.sqrt(0.8),T+10)

            v = []
            v += [eps_1[0]]
            for j in range(1,T+10):
                v += [eps_1[j] + 0.5*eps[j-1]]

            sigma = sigma_u
            u = sigma*np.array(v[10:])
            X1 = np.array(X1[10:])
            y0 = 0*X1+u 
            y1 = 1*X1+u


            y = np.hstack([y0[0:T//2],y1[T//2:T]])
            df['x_%s'%i] = X1
            df['y_%s'%i] = y
        df.to_csv('/Users/username/Dropbox/MIP_regr/MIP_regr_500/one_break/%s.csv'%fname)
```

```python
#DGP6
#one break
for sigma_u in [0.5, 1.0, 1.5]:
    for T in [100,200,500]:
        df = pd.DataFrame()
        fname = 'DGP-6_%s_%s'%(sigma_u,T)
        for i in tqdm(range(500)):
            np.random.seed(i+10*T+50000)
            
            u = np.random.normal(0,sigma_u,T+10)
            X1 = [0.2*u[0]]
            y = [0.2*X1[0] + u[0]]

            for j in range(1,10):
                beta = 0.2                
                X1 += [y[j-1]]
                y += [beta*X1[j]+u[j]]
            
            X1 = X1[-1:]
            y = y[-1:]
            
            for j in range(1,T):
                beta = 0.2
                if j >= T//2:
                    beta = 0.8
                
                X1 += [y[j-1]]
                y += [beta*X1[j]+u[j]]

            y = np.array(y)
            df['x_%s'%i] = X1
            df['y_%s'%i] = y
        df.to_csv('/Users/username/Dropbox/MIP_regr/MIP_regr_500/one_break/%s.csv'%fname)
```

```python
#many breaks 1
delta = 30
R = np.array([6, 10, 20])
for r in R:
    for sigma_u in [0.2, 0.5]:
        fname = 'DGP-1_%s_%s'%(sigma_u,r)
        df = pd.DataFrame()
        for i in range(500):
            np.random.seed(i+r*10000)
            

            L = r//2
            betas = [0]*delta + [1]*delta
            betas = betas*L
            u = np.random.normal(0,sigma_u,len(betas))
            X = np.random.normal(0,1,len(betas))
            Y = np.array(betas)*X + u
            
            df['x_%s'%i] = np.array(X)
            df['y_%s'%i] = np.array(Y)
        
        df.to_csv('/Users/username/Dropbox/MIP_regr/500_sim/many_breaks/%s.csv'%fname)
        

```
